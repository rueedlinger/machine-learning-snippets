# Clustering with K-means 

```python
%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd

from sklearn import datasets
from sklearn import cluster
from sklearn import manifold
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['label'] = iris.target
df.head()
```

Reduce the dataset to 2-dimension with MDS for better visualization

```python
mds = manifold.MDS(n_components=2)
new_dim = mds.fit_transform(iris.data)
df = pd.DataFrame(new_dim, columns=['X', 'Y'])
df.head()
```

Clustering with 2 clusters

```python
kmeans = cluster.KMeans(n_clusters=2)
label = kmeans.fit_predict(iris.data)
df['label'] = label

fig = plt.figure()
fig.suptitle('K-means k=2', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)

plt.scatter(df[df.label == 0].X, df[df.label == 0].Y, color='red', label='0')
plt.scatter(df[df.label == 1].X, df[df.label == 1].Y, color='blue', label='1')

plt.legend(bbox_to_anchor=(1.25, 1))
```

Clustering with 3 clusters

```python
kmeans = cluster.KMeans(n_clusters=3)
label = kmeans.fit_predict(iris.data)
df['label'] = label

fig = plt.figure()
fig.suptitle('K-means k=3', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)

plt.scatter(df[df.label == 0].X, df[df.label == 0].Y, color='red', label='0')
plt.scatter(df[df.label == 1].X, df[df.label == 1].Y, color='blue', label='1')
plt.scatter(df[df.label == 2].X, df[df.label == 2].Y, color='green', label='2')

plt.legend(bbox_to_anchor=(1.25, 1))
```

Clustering with 4 clusters

```python
kmeans = cluster.KMeans(n_clusters=4)
label = kmeans.fit_predict(iris.data)
df['label'] = label
df['label'] = label

fig = plt.figure()
fig.suptitle('K-means k=4', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)

plt.scatter(df[df.label == 0].X, df[df.label == 0].Y, color='red', label='0')
plt.scatter(df[df.label == 1].X, df[df.label == 1].Y, color='blue', label='1')
plt.scatter(df[df.label == 2].X, df[df.label == 2].Y, color='green', label='2')
plt.scatter(df[df.label == 3].X, df[df.label == 3].Y, color='yellow', label='3')

plt.legend(bbox_to_anchor=(1.25, 1))
```
